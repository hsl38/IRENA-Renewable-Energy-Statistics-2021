```python
import os
import pandas as pd
import numpy as np
```

```python
os.getcwd()
```

```python
oecd_pop_file_name = '..\\data\\oecd_population_1950_2020.xlsx'
df = pd.read_excel(oecd_pop_file_name, skiprows=4, skipfooter=1)
```

```python
df
```

```python
# clean country
df.rename({'Time' : 'country'}, axis=1, inplace=True)
```

```python
df.loc [~df['Unnamed: 1'].isna(), 'country'] = df.loc[~df['Unnamed: 1'].isna(), 'Unnamed: 1']
df
```

```python
for i in range(3, len(df.columns) - 1):
    if i % 2 == 1:
        # print(i, df.columns[i], i + 1, df.columns[i + 1])
        df[df.columns[i]] = df[df.columns[i + 1]]
df
```

```python
columns_to_drop = []
for col in df.columns:
    if 'Unnamed' in col:
        columns_to_drop.append(col)
print(columns_to_drop)
```

```python
df.drop(columns_to_drop, axis=1, inplace=True)
df
```

```python
df.dropna(axis=0, inplace=True)
df
```

```python
new_oecd_pop_file_name = '..\\data\\oecd_population_1950_2020_clean.xlsx'
df.to_excel(new_oecd_pop_file_name, index=False)
```
