```python
import os
import pandas as pd
import numpy as np
```

```python
os.getcwd()
```

```python
oecd_pop_file_name = '..\\data\\oecd_population_1950_2020_clean.xlsx'
```

```python
df_pop = pd.read_excel(oecd_pop_file_name)
df_pop
```

```python
filt = df_pop[df_pop['country'].isin(['G20', 'OECD - Total', 'World'])].index
filt
```

```python
df_pop.drop(filt, axis=0, inplace=True)
```

```python
df_pop.set_index('country', inplace=True)
```

```python
df_pop.columns
```

```python
gen_file_name = '..\\data\\irena_elec_2021_01.xlsx'
df_gen = pd.read_excel(gen_file_name)
df_gen
```

```python
df_cy = df_gen.groupby(['country', 'year'])[['capacity', 'generation']].sum()
```

```python
df_cy
```

```python
df_pop.index.tolist()
```

```python
df_cy['pop'] = 0
for i, r  in df_cy.iterrows():
    country, year = i
    # print(country, year)
    if country in df_pop.index.tolist():
        print(country, year, df_pop.loc[country, str(year)])
        df_cy.loc[i, 'pop'] = df_pop.loc[country, str(year)]
    else:
        # print(country, year)
        df_cy.loc[i, 'pop'] = np.NaN
```

```python
df_cy
```

```python
df_cy.dropna(axis=0, inplace=True)
df_cy
```

```python
df_cy['capacity_per_capita'] = df_cy['capacity'] / df_cy['pop'] * 1e3   # kW per person
df_cy['generation_per_capita'] = df_cy['generation'] / df_cy['pop'] * 1e6 # MWh per person
df_cy
```

```python
df_cy.reset_index(inplace=True)
df_cy
```

```python
df_cy.to_excel('2021_elec_per_capita.xlsx', index=False)
```
